Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

**Examples:**
Input: list1 = 5->2->3->8
list2 = 1->7->4->5
Output: New list = 5->7->4->8

Input:list1 = 2->8->9->3
list2 = 5->3->6->4
Output: New list = 5->8->9->4
</aside>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def create_linked_list(arr):
    if len(arr) == 0:
        return None

    nodes = [Node(data) for data in arr]

    head = nodes[0]
    curr = head

    for i in range(1, len(nodes)):
        curr.next = nodes[i]
        curr = curr.next

    return head

def print_linked_list(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

def merge_lists(list1, list2):
    if list1 is None:
        return list2
    if list2 is None:
        return list1

    curr1 = list1
    curr2 = list2
    new_head = None
    new_curr = None

    while curr1 and curr2:
        if curr1.data >= curr2.data:
            new_node = Node(curr1.data)
            curr1 = curr1.next
        else:
            new_node = Node(curr2.data)
            curr2 = curr2.next

        if new_head is None:
            new_head = new_node
            new_curr = new_node
        else:
            new_curr.next = new_node
            new_curr = new_curr.next

    while curr1:
        new_node = Node(curr1.data)
        new_curr.next = new_node
        new_curr = new_curr.next
        curr1 = curr1.next

    while curr2:
        new_node = Node(curr2.data)
        new_curr.next = new_node
        new_curr = new_curr.next
        curr2 = curr2.next

    return new_head

# Example 1
# list1: 5->2->3->8
# list2: 1->7->4->5
list1 = create_linked_list([5, 2, 3, 8])
list2 = create_linked_list([1, 7, 4, 5])

print("List 1:")
print_linked_list(list1)
print("List 2:")
print_linked_list(list2)

new_list = merge_lists(list1, list2)
print("New list:")
print_linked_list(new_list)

# Example 2
# list1: 2->8->9->3
# list2: 5->3->6->4
list1 = create_linked_list([2, 8, 9, 3])
list2 = create_linked_list([5, 3, 6, 4])

print("List 1:")
print_linked_list(list1)
print("List 2:")
print_linked_list(list2)

new_list = merge_lists(list1, list2)
print("New list:")
print_linked_list(new_list)
```

ðŸ’¡ **Question 2**

Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

**Example 1:**
Input:
LinkedList: 
11->11->11->21->43->43->60
Output:
11->21->43->60
    
**Example 2:**
Input:
LinkedList: 
10->12->12->25->25->25->34
Output:
10->12->25->34


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def create_linked_list(arr):
    if len(arr) == 0:
        return None

    nodes = [Node(data) for data in arr]

    head = nodes[0]
    curr = head

    for i in range(1, len(nodes)):
        curr.next = nodes[i]
        curr = curr.next

    return head

def print_linked_list(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

def remove_duplicates(head):
    if head is None or head.next is None:
        return head

    curr = head
    while curr.next:
        if curr.data == curr.next.data:
            curr.next = curr.next.next
        else:
            curr = curr.next

    return head

# Example 1
# LinkedList: 11->11->11->21->43->43->60
linked_list1 = create_linked_list([11, 11, 11, 21, 43, 43, 60])
print("Input:")
print_linked_list(linked_list1)

new_list1 = remove_duplicates(linked_list1)
print("Output:")
print_linked_list(new_list1)

# Example 2
# LinkedList: 10->12->12->25->25->25->34
linked_list2 = create_linked_list([10, 12, 12, 25, 25, 25, 34])
print("Input:")
print_linked_list(linked_list2)

new_list2 = remove_duplicates(linked_list2)
print("Output:")
print_linked_list(new_list2)
```

ðŸ’¡ **Question 3**

Given a linked list of sizeÂ **N**. The task is to reverse everyÂ **k**Â nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple ofÂ *k*Â then left-out nodes, in the end, should be considered as a group and must beÂ reversed (See Example 2 for clarification).

**Example 1:**
Input:
LinkedList: 1->2->2->4->5->6->7->8
K = 4
Output:4Â 2Â 2Â 1Â 8Â 7Â 6Â 5
Explanation:
The first 4 elements 1,2,2,4 are reversed first
and then the next 4 elements 5,6,7,8. Hence, the
resultant linked list is 4->2->2->1->8->7->6->5.                        
**Example 2:**
Input:
LinkedList: 1->2->3->4->5
K = 3
Output:3 2 1 5 4
Explanation:
The first 3 elements are 1,2,3 are reversed
first and then elements 4,5 are reversed.Hence,
the resultant linked list is 3->2->1->5->4.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def create_linked_list(arr):
    if len(arr) == 0:
        return None

    nodes = [Node(data) for data in arr]

    head = nodes[0]
    curr = head

    for i in range(1, len(nodes)):
        curr.next = nodes[i]
        curr = curr.next

    return head

def print_linked_list(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

def reverse_k_nodes(head, k):
    if head is None or head.next is None or k == 1:
        return head

    prev = None
    curr = head
    count = 0

    while curr and count < k:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
        count += 1

    if next_node:
        head.next = reverse_k_nodes(next_node, k)

    return prev

def reverse_linked_list_in_groups(head, k):
    if head is None or head.next is None or k == 1:
        return head

    count = 0
    curr = head

    while curr:
        count += 1
        if count == k:
            break
        curr = curr.next

    if count < k:
        return head

    next_group_head = reverse_linked_list_in_groups(curr.next, k)
    curr.next = None
    head = reverse_k_nodes(head, k)
    curr = head

    while curr.next:
        curr = curr.next

    curr.next = next_group_head

    return head

# Example 1
# LinkedList: 1->2->2->4->5->6->7->8
linked_list1 = create_linked_list([1, 2, 2, 4, 5, 6, 7, 8])
print("Input:")
print_linked_list(linked_list1)

reversed_list1 = reverse_linked_list_in_groups(linked_list1, 4)
print("Output:")
print_linked_list(reversed_list1)

# Example 2
# LinkedList: 1->2->3->4->5
linked_list2 = create_linked_list([1, 2, 3, 4, 5])
print("Input:")
print_linked_list(linked_list2)

reversed_list2 = reverse_linked_list_in_groups(linked_list2, 3)
print("Output:")
print_linked_list(reversed_list2)
```

ðŸ’¡ **Question 4**

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.

**Example:**

Inputs:   1->2->3->4->5->6->7->8->9->NULL and k = 3                    
Output:   3->2->1->4->5->6->9->8->7->NULL.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def create_linked_list(arr):
    if len(arr) == 0:
        return None

    nodes = [Node(data) for data in arr]

    head = nodes[0]
    curr = head

    for i in range(1, len(nodes)):
        curr.next = nodes[i]
        curr = curr.next

    return head

def print_linked_list(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

def reverse_alternate_k_nodes(head, k):
    if head is None or head.next is None or k == 1:
        return head

    prev = None
    curr = head
    count = 0

    while curr and count < k:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
        count += 1

    if next_node:
        head.next = next_node

    if count < k:
        return prev

    count = 0
    while curr and count < k - 1:
        curr = curr.next
        count += 1

    if curr:
        curr.next = reverse_alternate_k_nodes(curr.next, k)

    return prev

# Example
# LinkedList: 1->2->3->4->5->6->7->8->9->NULL
linked_list = create_linked_list([1, 2, 3, 4, 5, 6, 7, 8, 9])
print("Input:")
print_linked_list(linked_list)

reversed_list = reverse_alternate_k_nodes(linked_list, 3)
print("Output:")
print_linked_list(reversed_list)
```

ðŸ’¡ **Question 5**

Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.

**Examples**:

Input:   1->2->3->5->2->10, key = 2                         
Output:  1->2->3->5->10

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def create_linked_list(arr):
    if len(arr) == 0:
        return None

    nodes = [Node(data) for data in arr]

    head = nodes[0]
    curr = head

    for i in range(1, len(nodes)):
        curr.next = nodes[i]
        curr = curr.next

    return head

def print_linked_list(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

def delete_last_occurrence(head, key):
    if head is None:
        return None

    prev = None
    last_occurrence = None
    curr = head

    while curr:
        if curr.data == key:
            last_occurrence = curr
        curr = curr.next

    if last_occurrence:
        curr = head
        while curr.next != last_occurrence:
            curr = curr.next
        if prev:
            prev.next = last_occurrence.next
        else:
            head = last_occurrence.next

    return head

# Example
# LinkedList: 1->2->3->5->2->10
linked_list = create_linked_list([1, 2, 3, 5, 2, 10])
print("Input:")
print_linked_list(linked_list)

key = 2
updated_list = delete_last_occurrence(linked_list, key)
print("Output:")
print_linked_list(updated_list)
```

ðŸ’¡ **Question 6**

Given two sorted linked lists consisting ofÂ **N**Â andÂ **M**Â nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

**Examples:**

Input:Â a: 5->10->15,Â b: 2->3->20

Output:Â 2->3->5->10->15->20

Input:Â a: 1->1, b: 2->4

Output:Â 1->1->2->4


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def create_linked_list(arr):
    if len(arr) == 0:
        return None

    nodes = [Node(data) for data in arr]

    head = nodes[0]
    curr = head

    for i in range(1, len(nodes)):
        curr.next = nodes[i]
        curr = curr.next

    return head

def print_linked_list(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

def merge_sorted_lists(a, b):
    dummy = Node(None)
    curr_merged = dummy

    currA = a
    currB = b

    while currA and currB:
        if currA.data <= currB.data:
            curr_merged.next = currA
            currA = currA.next
        else:
            curr_merged.next = currB
            currB = currB.next
        curr_merged = curr_merged.next

    if currA:
        curr_merged.next = currA
    if currB:
        curr_merged.next = currB

    return dummy.next

# Example 1
# List A: 5->10->15
# List B: 2->3->20
a = create_linked_list([5, 10, 15])
b = create_linked_list([2, 3, 20])
print("Input:")
print("List A:")
print_linked_list(a)
print("List B:")
print_linked_list(b)

merged = merge_sorted_lists(a, b)
print("Output:")
print_linked_list(merged)

# Example 2
# List A: 1->1
# List B: 2->4
a = create_linked_list([1, 1])
b = create_linked_list([2, 4])
print("Input:")
print("List A:")
print_linked_list(a)
print("List B:")
print_linked_list(b)

merged = merge_sorted_lists(a, b)
print("Output:")
print_linked_list(merged)
```

ðŸ’¡ **Question 7**

Given aÂ **Doubly Linked List**, the task is to reverse the given Doubly Linked List.

**Example:**

Original Linked list 10 8 4 2            
Reversed Linked list 2 4 8 10

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def create_linked_list(arr):
    if len(arr) == 0:
        return None

    nodes = [Node(data) for data in arr]

    head = nodes[0]
    curr = head

    for i in range(1, len(nodes)):
        curr.next = nodes[i]
        nodes[i].prev = curr
        curr = curr.next

    return head

def print_linked_list(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

def reverse_doubly_linked_list(head):
    curr = head
    prev = None

    while curr:
        # Swap prev and next pointers
        next_node = curr.next
        curr.next = prev
        curr.prev = next_node

        # Move to the next node
        prev = curr
        curr = next_node

    # Update the head to point to the last node
    head = prev

    return head

# Example
# Original Linked list: 10 8 4 2
head = create_linked_list([10, 8, 4, 2])
print("Original Linked list:")
print_linked_list(head)

reversed_head = reverse_doubly_linked_list(head)
print("Reversed Linked list:")
print_linked_list(reversed_head)
```

ðŸ’¡ **Question 8**

Given a doubly linked list and a position. The task is to delete a node from given position inÂ a doubly linked list.

**Example 1:**           

Input:
LinkedList = 1 <--> 3 <--> 4
x = 3
Output:1 3
Explanation:After deleting the node at
position 3 (position starts from 1),
the linked list will be now as 1->3.

                   
**Example 2:**             
Input:
LinkedList = 1 <--> 5 <--> 2 <--> 9
x = 1
Output:5 2 9

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def create_doubly_linked_list(arr):
    if len(arr) == 0:
        return None

    nodes = [Node(data) for data in arr]

    head = nodes[0]
    curr = head

    for i in range(1, len(nodes)):
        curr.next = nodes[i]
        nodes[i].prev = curr
        curr = curr.next

    return head

def print_doubly_linked_list(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

def delete_node_at_position(head, position):
    if not head:
        return head

    # Case 1: Deleting the head node
    if position == 1:
        new_head = head.next
        if new_head:
            new_head.prev = None
        return new_head

    curr = head
    count = 1

    # Traverse to the node at the given position
    while curr and count < position:
        curr = curr.next
        count += 1

    # Case 2: Deleting a node in the middle
    if curr.next:
        curr.next.prev = curr.prev

    # Case 3: Deleting the tail node
    if curr.prev:
        curr.prev.next = curr.next

    return head

# Example 1
# Input: LinkedList = 1 <--> 3 <--> 4, position = 3
head = create_doubly_linked_list([1, 3, 4])
print("Original LinkedList:")
print_doubly_linked_list(head)

new_head = delete_node_at_position(head, 3)
print("LinkedList after deletion:")
print_doubly_linked_list(new_head)

# Example 2
# Input: LinkedList = 1 <--> 5 <--> 2 <--> 9, position = 1
head = create_doubly_linked_list([1, 5, 2, 9])
print("Original LinkedList:")
print_doubly_linked_list(head)

new_head = delete_node_at_position(head, 1)
print("LinkedList after deletion:")
print_doubly_linked_list(new_head)
```
